```python
import scipy
import numpy
import sklearn
import pandas
from sklearn.feature_selection import VarianceThreshold
from sklearn.tree import DecisionTreeClassifier
```

```python
matrix = '/data/projects/kimona/data/matrix_from_mcl.sorted.cutted.tsv'
names = '/data/projects/kimona/data/matrix_from_mcl.list'
labels_file = '/data/projects/kimona/data/matrix_from_mcl.labels'
```

```python
features = numpy.loadtxt(matrix)
print(features, features.shape)
```

```python
with open(labels_file) as f:
    lines = f.readlines()
    labels = []
    for line in lines:
        label = int(line.split()[1])
        if label == 0 or label == 1:
            labels.append(0)
        else:
            labels.append(1)

print(labels, len(labels))
```

```python
sel = VarianceThreshold(threshold=(.95 * (1 - .95)))
features2 = sel.fit_transform(features)
print(features2, features2.shape)
```

```python
clf = DecisionTreeClassifier()
clf.fit(features2, labels)
```

```python
import graphviz 
from sklearn import tree
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.format = 'svg'
graph.render()
```
